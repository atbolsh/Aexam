Here, we will plot the nonlinearities commonly used in neural networks, for the sake of section 2.

```python
import numpy as np
import torch
import torch.nn.functional as F

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm

x = np.arange(-10, 10, 0.01)

x_ = torch.FloatTensor(x)
sig_ = F.sigmoid(x_)
ReLU_ = F.relu(x_)
ELU_ = F.elu(x_)

sig = sig_.detach().numpy()
ReLU = ReLU_.detach().numpy()
ELU = ELU_.detach().numpy()
```

```python
fig=plt.figure()

plt.grid(True)

plt.plot(x, sig, 'b-')
plt.xlabel("x")
plt.ylabel("sigmoid")
plt.title("Sigmoid Nonlinearity")

#plt.show()
fig
```

```python
fig=plt.figure()

plt.grid(True)

plt.plot(x, ReLU, 'b-')
plt.xlabel("x")
plt.ylabel("ReLU")
plt.title("ReLU Nonlinearity")

#plt.show()
fig
```

```python
fig=plt.figure()

plt.grid(True)

plt.plot(x, ELU, 'b-')
plt.xlabel("x")
plt.ylabel("ELU")
plt.title("ELU Nonlinearity")

#plt.show()
fig
```
